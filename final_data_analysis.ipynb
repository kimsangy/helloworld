```python
import requests
import re
import pandas as pd
import matplotlib.pyplot as plt


pd.options.display.float_format = '{:.5f}'.format
pd.set_option('mode.chained_assignment',  None)

  
def get_dataset1(filename1):    #Dataset1 : Participating_Countries/2020,2021 GDP per capita 
    csv = pd.read_csv(filename1)
 #   print(csv)
 #   csv.fillna(0)
    list1 = []
    list2 = []
    list3 = []
    list4 = []
    list5 = []
    
    for i in range(len(csv)):
        if csv['2020_GDP'][i] != 0:
            if csv['NUM'][i] <=1:
                list1.append(csv['2020_GDP'][i])
            elif csv['NUM'][i] > 1 and csv['NUM'][i] <= 5:
                list2.append(csv['2020_GDP'][i])                
            elif csv['NUM'][i] > 5 and csv['NUM'][i] <= 10:
                list3.append(csv['2020_GDP'][i])          
            elif csv['NUM'][i] > 10 and csv['NUM'][i] <= 15:
                list4.append(csv['2020_GDP'][i])   
            elif csv['NUM'][i] > 15:
                list5.append(csv['2020_GDP'][i])   

    newlist1 = [x for x in list1 if pd.isnull(x) == False]
    newlist2= [x for x in list2 if pd.isnull(x) == False]
    newlist3= [x for x in list3 if pd.isnull(x) == False]
    newlist4= [x for x in list4 if pd.isnull(x) == False]
    newlist5= [x for x in list5 if pd.isnull(x) == False]

  #  print(newlist1)
  #  print(newlist2)
  #  print(newlist3)
  #  print(newlist4)    
  #  print(newlist5)  

    average_GDP_group1 = sum(newlist1)/len(newlist1)        # range1: Number of participations: 1
    average_GDP_group2 = sum(newlist2)/len(newlist2)        # range2: Number of participations 2~5
    average_GDP_group3 = sum(newlist3)/len(newlist3)        # range3: Number of participations 6~10
    average_GDP_group4 = sum(newlist4)/len(newlist4)        # range4: Number of participations 11~15
    average_GDP_group5 = sum(newlist5)/len(newlist5)        # range4: Number of participations over 16

    Num_of_paticipations = ['1', '2-5', '6-10', '11-15','over 16']
    average_GDP = [average_GDP_group1,average_GDP_group2,average_GDP_group3,average_GDP_group4,average_GDP_group5]


#    print(Num_of_paticipations)
#    print(average_GDP)

    plt.title("Correlation between the number of World Cup participations and GDP")
    plt.xlabel("Number of_paticipations")    
    plt.ylabel("Average GDP per capita (USD)")
    plt.bar(Num_of_paticipations, average_GDP)

 #   plt.show()




def get_dataset2(filename1):    #Dataset2 : Host_Countries/ GDP per capita of before and after year  
    csv = pd.read_csv(filename1)

    csv['YEAR(TEAM_CODE)'] = 0    


    for i in range(len(csv)):
        csv['YEAR(TEAM_CODE)'][i] = str(csv['HOST_YEAR'][i]) +  "(" + str(csv['TEAM_CODE'][i]) + ")"

 #   print(csv)

    csv.plot(x = 'YEAR(TEAM_CODE)', y =['BEFORE_YEAR_GDP', 'AFTER_YEAR_GDP'], kind="bar", rot =0, figsize = (20,10))

    plt.xlabel("Host year(Nation)")    
    plt.ylabel("GDP per capita (USD)")
    plt.xticks(rotation=90)
  #  plt.show()

if __name__ == '__main__':

    filename1 = "output_data/dataset1_Participating_Countries_GDP.csv"                            
    filename2 = "output_data/dataset2_Hosting_Countries_GDP.csv"                           
                                                                            

    get_dataset1(filename1)
    get_dataset2(filename2)



```
